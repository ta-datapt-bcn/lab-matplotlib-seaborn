#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8))
ax.plot(x, x**2, color ='red',label='y=x**2')
ax.plot(x, np.exp(x), color='blue', label='y=e**x')
ax.set_yscale('log')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Work_or_Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fitbit.head(2)
```

```python
# MEDIAN STEPS BY MONTH_ENCODED:
avg_steps_month = fitbit['Steps'].groupby(fitbit['Months_encoded']).mean()
labels1 = fitbit['Months'].sort_values().unique()
from sorted_months_weekdays import Month_Sorted_Month
labels1 = Month_Sorted_Month(labels1)
                             
# MEDIAN STEPS BY WORK_OR_WEEKEND:
work_weekend = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).mean() 
labels2 = ['Weekend', 'Week'] # weekend(0), work(1)

#PLOT
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
ax[0].plot(avg_steps_month.index, avg_steps_month, color='red')
ax[0].fill_between(avg_steps_month.index, avg_steps_month, alpha=0.2, color='royalblue')
ax[0].set_xticks(range(1,13))
ax[0].set_xticklabels(labels1, rotation=75)
ax[1].pie(work_weekend, labels=labels2, autopct='%1.2f%%', colors=['brown','darkblue'])
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4))
ls = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
r = 0
for i in ls:
    axes[r].scatter(fitbit[i], fitbit.Steps)
    r+=1
   
```
