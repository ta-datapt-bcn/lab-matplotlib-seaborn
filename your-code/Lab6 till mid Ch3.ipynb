#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(15,8))

ax[0].plot(x, x**2, "y", x, np.exp(x), "p")
ax[0].set_title("Normal scale")

ax[1].plot(x, x**2, "o", x, np.exp(x), "r")
ax[1].set_yscale("log")
ax[1].set_title("Logarithmic scale")

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
f, [ax1, ax2] = plt.subplots(1,2, figsize = (12,4))

# MEDIAN STEPS BY MONTH_ENCODED
plt.sca (ax1)

steps_month_week = fitbit["Steps"].groupby(fitbit["Months_encoded"]).median()

steps_month_week.plot(kind = "line")

plt.ylabel("Median number of steps")
plt.xlabel("Months")
plt.title("Median Steps per month")
plt.xticks(range(13), ["", "Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])

plt.savefig("figures/steps_per_month.png")

# MEDIAN STEPS BY WORK_OR_WEEKEND
plt.sca (ax2)

steps_month_week = fitbit["Steps"].groupby(fitbit["Work_or_Weekend"]).median()

steps_month_week.plot(kind = "area")

plt.ylabel("Median steps")
plt.xlabel("")
plt.title("Median Steps per Workdays and Weekend")
plt.xticks(range(2), ["Weekend","Workdays"])

plt.savefig("figures/steps_day_type.png")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
cols = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary", "Steps"]

fitbit_cols = fitbit[cols]

f, ax = plt.subplots(1, 3, figsize = (20,8))

for index in range(3):
    ax[index].scatter(fitbit_cols[cols[index]], fitbit_cols[cols[3]])
    ax[index].set_xlabel(fitbit_cols[cols[index]].name)
    ax[index].set_ylabel("Steps")
```

#### What additional information does the last plot provide about feature `Age`?

The violin plots combine features from boxplots and a density plots. The violin plot shows KDE and the thistribution of the data simultaneously, being the black line the range from min to max, the thicker slice the IQR (box) and the white pint the median.
